```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from bert_sklearn import BertClassifier
```

```python
train_df = pd.read_csv('https://raw.githubusercontent.com/aatishsuman/health-advice/master/data/pubmed_causal_language_use.csv')
train_df = train_df.sample(frac=1, random_state=24).reset_index(drop=True)
train_df
```

```python
test_df = pd.read_csv('https://raw.githubusercontent.com/aatishsuman/health-advice/master/data/test.csv')
test_df
```

```python
model = BertClassifier(random_state=42)
print(model)
model.fit(train_df['sentence'], train_df['label'])
```

```python
model = BertClassifier(bert_model='biobert-base-cased', random_state=42)
print(model)
model.fit(train_df['sentence'], train_df['label'])
```

```python
train_pred = model.predict(train_df['sentence'])
print('Train classification report:\n', classification_report(train_df['label'], train_pred, target_names=['0','1','2','3']), '\n')
```

```python
# Using all test data v1
print('Using all test data v1 - \n')

test_df['prediction'] = model.predict(test_df['sentence'])
print('Test classification report:\n', classification_report(test_df['label'], test_df['prediction'], target_names=['0','1','2','3']), '\n')
print('Confusion matrix -\n', confusion_matrix(test_df['label'], test_df['prediction'], labels=[0,1,2,3]))

error_df = test_df.query('label != prediction')
print('Total errors - ', error_df.shape[0], '\n')
print('Distribution of errors across type of study:\n', np.unique(error_df['design'], return_counts=True), '\n')
print('Distribution of type of relation across studies in test data:\n', pd.crosstab(test_df['design'], test_df['label']), '\n')
print('Distribution of type of relation across studies in errors:\n', pd.crosstab(error_df['design'], error_df['label']), '\n')
```

```python
# actual - no relation, predicted - correlation
list(error_df.query('label == 0 & prediction == 3')['sentence'])
```

```python
# actual - no relation, predicted - causation
list(error_df.query('label == 0 & prediction == 1')['sentence'])
```

```python
# Using all test data v2
print('Using all test data v2 - \n')

test_df_v2 = pd.read_csv('https://raw.githubusercontent.com/aatishsuman/health-advice/master/data/test_v2.csv')

test_df_v2['prediction'] = model.predict(test_df_v2['sentence'])
print('Test classification report:\n', classification_report(test_df_v2['label'], test_df_v2['prediction'], target_names=['0','1','2','3']), '\n')
print('Confusion matrix -\n', confusion_matrix(test_df_v2['label'], test_df_v2['prediction'], labels=[0,1,2,3]))

error_df = test_df_v2.query('label != prediction')
print('Total errors - ', error_df.shape[0], '\n')
print('Distribution of errors across type of study:\n', np.unique(error_df['design'], return_counts=True), '\n')
print('Distribution of type of relation across studies in test data:\n', pd.crosstab(test_df_v2['design'], test_df_v2['label']), '\n')
print('Distribution of type of relation across studies in errors:\n', pd.crosstab(error_df['design'], error_df['label']), '\n')
```

```python
# Using bottom half of articles for test data v2
print('Using bottom half for test data v2 - \n')

filtered_test_df = test_df_v2.query('location > count/2')
filtered_test_df['prediction'] = model.predict(filtered_test_df['sentence'])
print('Test classification report:\n', classification_report(filtered_test_df['label'], filtered_test_df['prediction'], target_names=['0','1','2','3']), '\n')
print('Confusion matrix -\n', confusion_matrix(filtered_test_df['label'], filtered_test_df['prediction'], labels=[0,1,2,3]))

error_df = filtered_test_df.query('label != prediction')
print('Total errors - ', error_df.shape[0], '\n')
print('Distribution of errors across type of study:\n', np.unique(error_df['design'], return_counts=True), '\n')
print('Distribution of type of relation across studies in test data:\n', pd.crosstab(filtered_test_df['design'], filtered_test_df['label']), '\n')
print('Distribution of type of relation across studies in errors:\n', pd.crosstab(error_df['design'], error_df['label']), '\n')
```
