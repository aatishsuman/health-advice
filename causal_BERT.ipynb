```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from bert_sklearn import BertClassifier
```

```python
train_df = pd.read_csv('https://raw.githubusercontent.com/aatishsuman/health-advice/master/data/pubmed_causal_language_use.csv')
train_df = train_df.sample(frac=1, random_state=24).reset_index(drop=True)
train_df
```

```python
test_df = pd.read_csv('https://raw.githubusercontent.com/aatishsuman/health-advice/master/data/test.csv')
test_df
```

```python
model = BertClassifier(random_state=42)
print(model)
model.fit(train_df['sentence'], train_df['label'])
```

```python
model = BertClassifier(bert_model='biobert-base-cased', random_state=42)
print(model)
model.fit(train_df['sentence'], train_df['label'])
```

```python
train_pred = model.predict(train_df['sentence'])
print('Train classification report:\n', classification_report(train_df['label'], train_pred, target_names=['0','1','2','3']), '\n')

test_pred = model.predict(test_df['sentence'])
print('Test classification report:\n', classification_report(test_df['label'], test_pred, target_names=['0','1','2','3']), '\n')
```

```python
test_df['prediction'] = test_pred
print('Confusion matrix -\n', confusion_matrix(test_df['label'], test_df['prediction'], labels=[0,1,2,3]))

error_df = test_df.query('label != prediction')
print('Total errors - ', error_df.shape[0], '\n')
print('Distribution of type of study:\n', np.unique(error_df['design'], return_counts=True), '\n')
```

```python
# actual - no relation, predicted - correlation
list(error_df.query('label == 0 & prediction == 3')['sentence'])
```
